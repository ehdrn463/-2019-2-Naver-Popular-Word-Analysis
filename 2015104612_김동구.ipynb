#### 네이버의 의제설정 효과는 약해졌다.
1. 의제설정 효과는 특정 매체에서 중요시 여기는 것을 시민들도 중요시 여기는 것이다.
2. 네이버의 실시간 검색어 정책 전환 전에는 광고 키워드가 분류되지 않았다.
  - 그렇다면 대부분의 실시간 검색어를 네이버 이용자들이 중요시 여겼다고 할 수 있다.
3. 네이버의 실시간 검색어 정책 전환 후에는 검색된 키워드 약 1,000개 중에 200개 정도가 광고 키워드로 분류되었다.
  - 이 광고 키워드는 에비어의 이용자들이 원하지 않는 정보임을 SNS를 통해 쉽게 확인해볼 수 있다.
  - 즉 의제설정 효과를 측정하는 변수를 실시간 검색어 갯수로 판단할 때, 20%정도 의제설정 효과가 약해졌다고 할 수 있다.

#### 네이버 데이터랩 API를 이용한 검색 트렌드 수집
  1. 설정한 기간 내에 대한 특정 키워드 검색 기록 수집
  2. 가장 많이 검색된 날짜의 데이터를 100으로 설정
  3. 검색 데이터를 0 ~ 100상대적인 데이터로 표시
 

```python
def getTrend(keyword):
    # api id, secret 값 받아서 입력해야 함
    client_id = ""
    client_secret = ""
    url = "https://openapi.naver.com/v1/datalab/search"

    body = {
        "startDate":"2017-01-10",
        "endDate":"2019-10-09",
        "timeUnit":"date",
        "keywordGroups":[
            {"groupName":keyword,"keywords":[keyword]},
        ],
    }
    
    body = json.dumps(body)

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    request.add_header("Content-Type","application/json")
    response = urllib.request.urlopen(request, data=body.encode("utf-8"))
    json_obj = json.loads(response.read().decode('utf-8'))

    df = pd.DataFrame(json_obj['results'][0]['data'])
    df.columns=['기간', keyword]
    return df
```

```python
# test_data: 기간 재설정된 실시간 검색어 데이터
# date_list: 날짜 데이터
def cumulating_trend(test_data, col_lists):
    # root 사용 이유: 1열 데이터가 찾으려는 기간동안 검색데이터가 존재해야지 다음 열이 알맞게 들어감
    root = '경희대'
    getTrendData = getTrend(root)
    root_df = pd.concat([getTrendData],axis=1)
    
    # 날짜 가져오기
    for col_list in col_lists:
        # 날짜 내 실시간 검색어 리스트 만들기
        col_keyword_lists = list(test_data.loc[:,col_list])
        for col_keyword in col_keyword_lists:
            getTrendData = getTrend(col_keyword)
            
            # 키워드가 없었으면 바로 keyword_box에 저장
            if col_keyword not in keyword_box:
                keyword_box.append(col_keyword)
            else:
                # numberK: 중복된 횟수
                numberK = keyword_box.count(col_keyword)
                keyword_box.append(col_keyword)
                col_keyword+=str(numberK)
                
            new_df = pd.concat([getTrendData], axis=1)
        
            if col_list == col_lists[0]:
                if col_keyword == col_keyword_lists[0]:
                    sum_df = root_df.join(new_df.set_index('기간'),on='기간')
                else:
                    sum_df = sum_df.join(new_df.set_index('기간'), on = '기간')              
            else :
                try:
                    sum_df = sum_df.join(new_df.set_index('기간'), on='기간')
                except:
                    # 중복이 여러 번이면.... 어떻게 만들지 고민해야 함.
                    new_df.rename(columns = {new_df.columns.values[1] : col_keyword}, inplace = True)
                    sum_df = sum_df.join(new_df.set_index('기간'), on='기간')       
                
    # root_df 지우기
    del sum_df['경희대']
    
    # 결측값 0으로 채우기 (이 데이터에선 결측값 = 0)
    sum_df = sum_df.fillna(0)
    return sum_df


# dataframe 파일로 저장
def saving_df(df):
    sum_df.to_csv("trend_filter/"+time_clock+"trend.csv", encoding='ms949')              
```
